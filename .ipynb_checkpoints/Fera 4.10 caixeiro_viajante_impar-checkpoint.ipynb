## 4.10 O caixeiro que prefere cidades ímpares

#### Objetivo:
Encontre o caminho de menor distância no problema do caixeiro viajante
que prefere cidades ímpares e mostre ele de forma gráfica.

**Considerações do experimento**: Considere um número n ≥ 7 de coordenadas
(x, y) de cidades (cada cidade ocupa uma posição (x, y) diferente). Você pode gerar as
coordenadas de forma aleatória ou simplesmente usar as coordenadas que desejar. O
caixeiro só anda em linha reta e apenas entre duas cidades. O caixeiro começa e termina
seu trajeto na mesma cidade e, fora a cidade inicial, ele não visita nenhuma outra cidade
mais de uma vez. Além disso, atribua um número inteiro para cada uma das n cidades que
o caixeiro irá visitar, iniciando a contagem pelo número zero e aumentando esse número
de 1 em 1. O caixeiro deverá necessariamente visitar primeiro as cidades com números
ímpares antes das cidades com números pares. A cidade de número zero deve ser a cidade
inicial.

## Introdução


O Problema do Caixeiro Viajante (TSP) clássico consiste em encontrar a rota de menor distância que passa por todas as cidades exatamente uma vez e retorna ao ponto de partida. Nesta variação proposta, o caixeiro começa e termina na cidade **0** e deve obrigatoriamente visitar **primeiro todas as cidades de índice ímpar** antes de partir para as cidades de índice par. Isso adiciona uma restrição de ordem que foge do problema tradicional e exige um tratamento especial no momento de construir e refinar cada rota.

Para lidar com essa restrição, implementamos um **Algoritmo Genético** que evolui uma população de soluções candidatas ao longo de várias gerações. Cada indivíduo é uma lista de cidades que representa um percurso. A cada geração, aplicamos:

1. **Seleção por Torneio**: sorteamos grupos pequenos de indivíduos e escolhemos o melhor de cada grupo para formar a base do cruzamento.  
2. **Cruzamento Ordenado**: trocamos segmentos entre pares de pais, mantendo a característica de permutação.  
3. **Mutação por Troca**: embaralhamos dois genes aleatórios em cada indivíduo, para introduzir diversidade.  
4. **Reparo “Ímpar → Par”**: após cada cruzamento ou mutação usamos uma função que recoloca a “Cidade 0” no início, seguido de todos os índices ímpares (na ordem quew estava no indíviduo) e por fim os pares.

Ao final do processo, coletamos o melhor trajeto de cada geração (nosso “hall da fama”) e exibimos graficamente aquele que apresenta a menor distância total, destacando como a restrição de visitas ímpares antes de pares foi cumprida.  

---


```python
from pprint import pprint
from functools import partial
from itertools import permutations
import matplotlib.pyplot as plt

from funcoes_caixeiro import cria_cidades
from funcoes_caixeiro import populacao_caixeiro as cria_populacao
from funcoes_caixeiro import funcao_objetivo_pop_caixeiro
from funcoes_caixeiro import funcao_objetivo_caixeiro
from funcoes_caixeiro import selecao_torneio_min as funcao_selecao
from funcoes_caixeiro import cruzamento_ordenado as funcao_cruzamento
from funcoes_caixeiro import mutacao_troca as funcao_mutacao
from funcoes_caixeiro import plota_trajeto
```

```python
NUM_CIDADES = 20
CIDADES = cria_cidades(NUM_CIDADES)

TAMANHO_POPULACAO = 100
NUM_GERACOES = 500
CHANCE_DE_CRUZAMENTO = 0.5
CHANCE_DE_MUTACAO = 0.05
TAMANHO_TORNEIO = 3
```

```python
populacao = cria_populacao(TAMANHO_POPULACAO, CIDADES)
# pprint(populacao)
# funcao_objetivo(populacao, CIDADES)
for i, j in enumerate(populacao):
    print(i,j)
```

```python
funcao_objetivo = partial(funcao_objetivo_pop_caixeiro, cidades=CIDADES)
CIDADES
```

```python
hall_da_fama = []

for n in range(NUM_GERACOES):
    
    # Seleção
    fitness = funcao_objetivo(populacao)        
    selecionados = funcao_selecao(populacao, fitness, TAMANHO_TORNEIO)
    
    # Cruzamento
    proxima_geracao = []
    for pai, mae in zip(selecionados[::2], selecionados[1::2]):
        individuo1, individuo2 = funcao_cruzamento(pai, mae, CHANCE_DE_CRUZAMENTO)
        proxima_geracao.append(individuo1)
        proxima_geracao.append(individuo2)
    
    # Mutação
    funcao_mutacao(proxima_geracao, CHANCE_DE_MUTACAO)
    
    # Atualização do hall da fama
    fitness = funcao_objetivo(proxima_geracao)
        
    menor_fitness = min(fitness)
    indice = fitness.index(menor_fitness)
    hall_da_fama.append(proxima_geracao[indice])   
    

    # Encerramento
    populacao = proxima_geracao

for indice, individuo in enumerate(hall_da_fama):
    if indice == 0 or individuo != hall_da_fama[indice - 1]:
        print(indice, individuo)
                
```

```python
fitness = funcao_objetivo(hall_da_fama)
menor_fitness = min(fitness)
indice = fitness.index(menor_fitness)
melhor_individuo_observado = hall_da_fama[indice]
distancia = funcao_objetivo_caixeiro(melhor_individuo_observado, CIDADES)

print()
print("Melhor individuo obtido por algoritmos genéticos:")
print(melhor_individuo_observado, "com distância:", distancia)
print()

plota_trajeto(CIDADES, melhor_individuo_observado)
```
