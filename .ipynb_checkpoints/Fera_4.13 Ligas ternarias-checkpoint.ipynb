```python
import random
from pprint import pprint
from functools import partial
from itertools import product

from funcoes_fera_13 import funcao_objetivo_pop_liga as funcao_objetivo
from funcoes_fera_13 import cria_populacao_liga as cria_populacao
from funcoes_fera_13 import selecao_torneio_max as funcao_selecao
from funcoes_fera_13 import mutacao_pesos_liga as mutacao_pesos
from funcoes_fera_13 import mutacao_elementos_liga as mutacao_elementos
from funcoes_fera_13 import cruzamento_uniforme_dicionario as funcao_cruzamento
from funcoes_fera_13 import calcula_liga as calcula_liga
```

```python
LIMITE = 100

TAMANHO_POPULACAO = 10
NUM_GERACOES = 100
CHANCE_DE_CRUZAMENTO = 0.5
CHANCE_DE_MUTACAO = 0.05
TAMANHO_LIGA = 3
```

```python
def funcao_objetivo_liga(candidato, limite):
    peso_total = 0
    valor_total = 0
    
    for elemento in candidato:
        peso_total += candidato[elemento]["Peso"]
        valor_total += candidato[elemento]["Valor"]
    if peso_total <= limite:
        return valor_total, peso_total
    else:
        return 0.01
    
funcao_objetivo_liga (populacao[1],100)
```

```python
def funcao_objetivo_pop_liga(populacao, limite):
    fitness = []
    
    for candidato in populacao:
        fitness.append(funcao_objetivo_liga(candidato, limite))
                    
    return fitness
fitness = funcao_objetivo_pop_liga(populacao, 100)
fitness
```

```python
def selecao_torneio_max(populacao, fitness, tamanho_torneio):
    """Faz a seleção de uma população usando torneio.

    Nota: da forma que está implementada, só funciona em problemas de
    maximização.

    Args:
      populacao: lista contendo os individuos do problema
      fitness: lista contendo os valores computados da funcao objetivo
      tamanho_torneio: quantidade de invíduos que batalham entre si

    """
    selecionados = []

    for _ in range(len(populacao)):
        sorteados = random.sample(populacao, tamanho_torneio)

        fitness_sorteados = []
        for individuo in sorteados:
            indice_individuo = populacao.index(individuo)
            fitness_sorteados.append(fitness[indice_individuo])

        max_fitness = max(fitness_sorteados)
        indice_max_fitness = fitness_sorteados.index(max_fitness)
        individuo_selecionado = sorteados[indice_max_fitness]

        selecionados.append(individuo_selecionado)

    return selecionados

selecionados = selecao_torneio_max(populacao, fitness, 3)
selecionados
```

```python
def cruzamento_uniforme_dicionario(pai, mae, chance_de_cruzamento):
    """Realiza cruzamento uniforme entre elementos, passando os valores associados a ele para os filhos

    Args:
      pai: dicionário representando um individuo
      mae: dicionário representando um individuo
      chance_de_cruzamento: float entre 0 e 1 representando a chance de cruzamento

    """
    if random.random() < chance_de_cruzamento:
        filho1 = {}
        filho2 = {}
           

        for gene_pai, gene_mae in zip(pai, mae):
            if random.choice([True, False]):
                filho1[gene_pai] = pai[gene_pai]
                filho2[gene_mae] = mae[gene_mae]
            else:
                    filho1[gene_mae] = mae[gene_mae]
                    filho2[gene_pai] = mae[gene_mae]
                      

        return filho1, filho2
    else:
        return pai, mae
    
cruzamento_uniforme_dicionario(selecionados[0], selecionados[1], 0.5)
```

```python

def mutacao_elementos_liga(populacao, chance_de_mutacao, elementos_possiveis):
    """Realiza mutação simples

    Args:
      populacao: lista contendo os indivíduos do problema
      chance_de_mutacao: float entre 0 e 1 representando a chance de mutação
      valores_possiveis: lista com todos os valores possíveis dos genes

    """
    for individuo in populacao:
        if random.random() < chance_de_mutacao:
            elementos = list(individuo.keys())
            elementos_possiveis_lista = list(elementos_possiveis.keys())
            elemento_mutado = random.sample(elementos,1)[0]
            elemento_novo = random.sample(elementos_possiveis_lista,1)[0]
            individuo[elemento_novo] = individuo.pop(elemento_mutado)
            individuo[elemento_novo]["Valor"] =  elementos_possiveis[elemento_novo]* individuo[elemento_novo]["Peso"]  
    return populacao
mutacao_elementos_liga(selecionados, 0.7, preco_elemento_grama)
```

```python
def mutacao_pesos_liga(populacao, chance_de_mutacao):
 
    """Aplica mutacao de troca em um indivíduo
 
    Args:
      populacao: lista contendo os indivíduos do problema
      chance_de_mutacao: float entre 0 e 1 representando a chance de mutação
 
    """
    for individuo in populacao:
        if random.random() < chance_de_mutacao:
            elementos = list(individuo.keys())
 
            elementos_usados = random.sample(elementos, 2)

            elemento_1 = elementos_usados[0]
            elemento_2 = elementos_usados[1]

            while individuo[elemento_1]["Peso"] < 10:
                elementos_usados = random.sample(elementos, 2)

                elemento_1 = elementos_usados[0]
                elemento_2 = elementos_usados[1]
 
            peso_removido = random.randint(1,5)
 
            individuo[elemento_1]["Peso"] = individuo[elemento_1]["Peso"] - peso_removido
            individuo[elemento_2]["Peso"] = individuo[elemento_2]["Peso"] + peso_removido                 
    return populacao
mutacao_pesos_liga(selecionados, 0.7)
```

```python
funcao_objetivo = partial(funcao_objetivo, 
                          limite=LIMITE)
```

```python
hall_da_fama = []

for n in range(NUM_GERACOES):
    
    # Seleção
    fitness = funcao_objetivo(populacao)        
    selecionados = funcao_selecao(populacao, fitness, 3)
    
    # Cruzamento
    proxima_geracao = []
    for pai, mae in zip(selecionados[::2], selecionados[1::2]):
        individuo1, individuo2 = funcao_cruzamento(pai, mae, CHANCE_DE_CRUZAMENTO, preco_elemento_grama)
        proxima_geracao.append(individuo1)
        proxima_geracao.append(individuo2)
    
    # Mutação
    mutacao_elementos(populacao, CHANCE_DE_MUTACAO, preco_elemento_grama)
    mutacao_pesos(populacao, CHANCE_DE_MUTACAO)
    
    # Atualização do hall da fama
    fitness = funcao_objetivo(proxima_geracao)
        
    maior_fitness = max(fitness)
    indice = fitness.index(maior_fitness)
    hall_da_fama.append(proxima_geracao[indice])    
    
    # Encerramento
    populacao = proxima_geracao
    
for indice, individuo in enumerate(hall_da_fama):
    if indice == 0 or individuo != hall_da_fama[indice - 1]:
        print(indice, individuo)
```
