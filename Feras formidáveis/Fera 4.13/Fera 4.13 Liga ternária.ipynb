Objetivo: Encontre uma liga de três elementos que tenha o maior custo possível. A
liga ternária deve ser da forma x A.y B.z C sendo que x + y + z = 100 g, x ≥ 5 g, y ≥ 5 g,
z ≥ 5 g e “A”, “B” e “C” são elementos químicos diferentes. Utilize o preço dado abaixo
[1]. Considere que qualquer composto com 3 elementos químicos é chamado de liga.
Dica: Pode ser interessante criar uma função que gere a população inicial para garantir
que x + y + z = 100 g.
Dica 2: Pode ser interessante criar uma ou mais funções de cruzamento e mutação
neste problema de forma que todas elas garantam que, ao final do processo do operador,
os indivíduos mantenham a característica de ter x + y + z = 100 g.

Estratégia:

Os operadores genéticos foram formulados a partir das condições de contorno. 

**1. Geração da população inicial** 
Cada indivíduo é representado por um dicionário onde as chaves são os nomes dos elementos químicos (genes) e os valores são dicionários contendo "Peso" (massa em gramas) e "Valor" (relacionado ao preço do elemento). A geração inicial assegura que cada elemento tenha pelo menos 5g e que a soma total dos pesos seja igual a 100g para a liga.

**2. Cruzamento uniforme personalizado**  
O cruzamento uniforme seleciona aleatoriamente para cada gene se o valor será herdado do pai ou da mãe. O filho 1 mantém o peso do pai para todos os genes, enquanto o filho 2 mantém o peso da mãe, alterando apenas os elementos e valores conforme o cruzamento. O método funciona mesmo quando pai e mãe têm números diferentes de genes.

**3. Mutação**
- Mutação de elementos: substitui aleatoriamente um elemento de um indivíduo por outro da lista de elementos possíveis, mantendo o peso original e atualizando o valor conforme o novo elemento.  
- Mutação de pesos: redistribui uma quantidade de peso entre dois elementos do mesmo indivíduo, removendo peso de um (com peso mínimo garantido) e adicionando ao outro, sem alterar os elementos em si. Depois os valores são atualizados conforme o novo peso.

**4. Função Objetivo**

Para a função objetivo, calculamos o valor total, que é correspondente a somatória dos pesos vezes o valor dos elementos em grama.  


**5. Manipulação de Dados**

```python
LIMITE = 100

TAMANHO_POPULACAO = 10
NUM_GERACOES = 1000
CHANCE_DE_CRUZAMENTO = 0.5
CHANCE_DE_MUTACAO = 0.05
TAMANHO_LIGA = 3
```

```python
populacao = cria_populacao(TAMANHO_POPULACAO, TAMANHO_LIGA, preco_elemento_grama)

populacao
```

```python
funcao_objetivo = partial(funcao_objetivo)
```

```python
hall_da_fama = []

for n in range(NUM_GERACOES):
    
    # Seleção
    fitness = funcao_objetivo(populacao)        
    selecionados = funcao_selecao(populacao, fitness, 3)
    
    # Cruzamento
    proxima_geracao = []
    for pai, mae in zip(selecionados[::2], selecionados[1::2]):
        individuo1, individuo2 = funcao_cruzamento(pai, mae, CHANCE_DE_CRUZAMENTO, preco_elemento_grama)
        proxima_geracao.append(individuo1)
        proxima_geracao.append(individuo2)
    
    # Mutação
    mutacao_elementos(populacao, CHANCE_DE_MUTACAO, preco_elemento_grama)
    mutacao_pesos(populacao, CHANCE_DE_MUTACAO, preco_elemento_grama)
    
    # Atualização do hall da fama
    fitness = funcao_objetivo(proxima_geracao)
        
    maior_fitness = max(fitness)
    indice = fitness.index(maior_fitness)
    hall_da_fama.append(proxima_geracao[indice])    
    
    # Encerramento
    populacao = proxima_geracao
```

```python


# Avaliar fitness de todos os melhores indivíduos
fitness = funcao_objetivo(hall_da_fama)

# Identificar o melhor fitness e o índice correspondente
maior_fitness = max(fitness)
indice_melhor = fitness.index(maior_fitness)

# Recuperar o melhor indivíduo
melhor_individuo = hall_da_fama[indice_melhor]

# Calcular peso e valor total (ou outra métrica)
peso_total, valor_total = calcula_liga(melhor_individuo)

print("Melhor indivíduo:", melhor_individuo)
print(f"Peso total: {peso_total:.2f}")
print(f"Valor total: {valor_total:.2f}")
```
